```python
from langchain_huggingface import HuggingFaceEmbeddings
import numpy as np
```

```python
resume_text = """
Johnathan M. Carter
Data Analyst with 3+ years of experience...

SKILLS:
Python, SQL, R, Pandas, NumPy, Scikit-learn, Power BI, Tableau, Machine Learning
"""

job_description = """
BlueWave Technologies is hiring a Data Analyst.

REQUIRED SKILLS:
Python, SQL, Tableau, Power BI, ETL, Machine Learning, Cloud (AWS/GCP)
"""
```

```python
# rumus similarity
# https://www.geeksforgeeks.org/python/how-to-calculate-cosine-similarity-in-python/

def cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
# get model
emb_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# ubah text jadi vectir
resume_vector = emb_model.embed_query(resume_text)
job_vector = emb_model.embed_query(job_description)

print(resume_vector)
print(job_vector)
```

```python
# hasil similarity
res = cosine_similarity(resume_vector, job_vector)
res
```
